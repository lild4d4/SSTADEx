```python
from sstadex import simplediffpair
import matplotlib.pyplot as plt
import numpy as np
```

```python
vs = np.linspace(0, 0.9, 100)
vds_diffpar = 1.2 - vs
vgs_diffpar = 0.9 - vs

parametros_electricos = {'Vds': vds_diffpar, 'Vsb': 0, 'Vgs': vgs_diffpar}

difpair_primitive = simplediffpair(netlist='simplediffpair.spice', type='nmos', inputs={'vds_lut': (0.1, 0.9, 0.01), 'vgs_lut': (0.1, 1.8, 0.01), 'vds': parametros_electricos['Vds'], 'vgs': parametros_electricos['Vgs'], 'il': 20e-6,
                  'length': [0.4e-06, 0.8e-06, 1.6e-06, 3.2e-06, 6.4e-6], '2d_var': ['vds', 'vgs']})
difpair_primitive.build()
```

```python
gain_difpair = 20*np.log10(difpair_primitive.Ro*difpair_primitive.gm)
plt.plot(gain_difpair)
```
