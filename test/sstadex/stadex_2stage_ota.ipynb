# First stage

```python
# OTA macro definition

OTA_macro = Macromodel(name = "ota_cap",
                       macromodel_parameters={Symbol('Ra'): np.linspace(100, 100000, 5),
                                              Symbol('gma'): np.linspace(0.001, 0.1, 5)},
                       req_tfs = [["vout", "vpos"], ["vout", "vpos"]],
                       electrical_parameters={Symbol("gdif_2"): Symbol("gdif_1"),
                                              Symbol("Rdif_2"): Symbol("Rdif_1")},
                        output = [Symbol('W'),
                                  Symbol('mesh')])

# Diffpair primitive definition

vs = np.linspace(0.1, 0.8, 20)
vds_diffpar = 1.2 - vs
vgs_diffpar = 0.9 - vs

parametros_electricos = {'Vds': vds_diffpar, 'Vsb': 0, 'Vgs': vgs_diffpar}

diffpair_primitive = simplediffpair(netlist='simplediffpair.spice', type='nmos', inputs={'vds_lut': (0.1, 1.2, 0.01), 'vgs_lut': (0.1, 1.8, 0.01), 'vds': parametros_electricos['Vds'], 'vgs': parametros_electricos['Vgs'], 'il': 20e-6,
                  'length': [0.4e-06, 0.8e-06, 1.6e-06, 3.2e-06, 6.4e-6], '2d_var': ['vds', 'vgs']})

diffpair_primitive.build()

diffpair_primitive.parameters = {Symbol("Rdif_1"): np.asarray(diffpair_primitive.Ro).flatten(), 
                                 Symbol("gdif_1"): np.asarray(diffpair_primitive.gm).flatten(),
                                 Symbol("Cgd"): np.asarray(diffpair_primitive.cgd).flatten(),
                                 Symbol("Cgs"): np.asarray(diffpair_primitive.cgs).flatten()}
diffpair_primitive.output = {Symbol('W'): np.asarray(diffpair_primitive.W).flatten(),
                             Symbol('mesh'): np.asarray(diffpair_primitive.mesh)}

# Active Load primitive definition

ota_cm_pmos = cm_pmos(name='ota_cm_pmos', netlist='pmos_cm.spice', type='pmos', inputs={'vds_lut': -0.6, 'vgs_lut': (-1.8, -0.1, 0.01), 'vgs': -0.6, 'il': 20e-6,
                  'length': [0.4e-06, 0.8e-06, 1.6e-06, 3.2e-06, 6.4e-6], '2d_var': ['length', 'vgs']})
ota_cm_pmos.build()

ota_cm_pmos.parameters = {Symbol("Raload_1"): np.asarray(ota_cm_pmos.Ro).flatten(),
                          Symbol("Raload_2"): 1/np.asarray(ota_cm_pmos.gm).flatten(),
                          Symbol("gaload"): np.asarray(ota_cm_pmos.gm).flatten()}

current_source = Macromodel(name = "currentSource")

# OTA macro submodules and primitives update

OTA_macro.submacromodels = [current_source]
OTA_macro.primitives = [diffpair_primitive, ota_cm_pmos]

class test:
    def __init__(self) -> None:
        pass
    def eval(funct):
        return {"eval": funct}

#################### PSRR #######################
psrr = test()
psrr.tf = ("vout", "vpos")
psrr.parametros = {Symbol("gdif_2"): Symbol("gdif_1"),
                   Symbol("Rdif_2"): Symbol("Rdif_1"),
                   Symbol("V1"): 0,
                   Symbol("V_n"): 0,
                   Symbol("V_p"): 1,
                   Symbol("I2"): 0,
                   Symbol("s"): 0,
                   Symbol("Cl"): 1e-12}

psrr.conditions = {"min": [40]}
psrr.variables = {}
psrr.out_def = {"eval": psrr.tf}

#################### GBW #######################

gbw = test()
gbw.tf = ("vout", "vpos")
gbw.parametros = {Symbol("gdif_2"): Symbol("gdif_1"),
                   Symbol("Rdif_2"): Symbol("Rdif_1"),
                   Symbol("V1"): 0,
                   Symbol("V_n"): 0,
                   Symbol("V_p"): 1,
                   Symbol("I2"): 0,
                   Symbol("Cl"): 1e-12}
gbw.variables = {}
gbw.out_def = {"eval": gbw.tf}

OTA_macro.especificaciones = [psrr, gbw]

results, axes, primoutputs = dfs(OTA_macro)
```

```python
W_diff = np.tile(diffpair_primitive.W.flatten(), 5).reshape(5, 100)
L_diff = np.tile(diffpair_primitive.L.flatten(), 5).reshape(5, 100)
WL_diff =W_diff+L_diff

W_al = np.repeat(ota_cm_pmos.W.flatten(), 100).reshape(5, 100)
L_al = np.repeat(ota_cm_pmos.L.flatten(), 100).reshape(5, 100)
WL_al = W_al+L_al

area_1stage = WL_diff+WL_al
gain_1stage = 20*np.log10(np.abs(results[0].reshape(5, 100)))
```

```python
gain_1stage_mask = np.full(gain_1stage.flatten().shape, True)
for idx, gain in enumerate(gain_1stage.flatten()):
    for cond in psrr.conditions["min"]:
        if gain>cond:
            gain_1stage_mask[idx] = True
        else:
            gain_1stage_mask[idx] = False
```

```python
plt.scatter(area_1stage.flatten(), gain_1stage.flatten())
plt.scatter(area_1stage.flatten()[gain_1stage_mask], gain_1stage.flatten()[gain_1stage_mask])
plt.xscale("log")
```

```python
frec = Symbol("frec")

gbw_j = []

for i in results[1][0][gain_1stage_mask]:
    temp = i.subs({Symbol("s"): frec*2j*np.pi})
    gbw_j.append(temp)
```

```python
f = np.logspace(1, 13, 200)
gbw_final = []
phase_final = []

for i in gbw_j:
    gbw_lamb = lambdify(frec, i)
    raw = gbw_lamb(f)
    abs_values = 20*np.log10(np.abs(raw))
    abs_values = abs_values-3
    cross = np.argmin(np.abs(abs_values))
    gbw_final.append(f[cross])
    phase_final.append(np.angle(raw[cross])*180/(np.pi))
```

```python
plt.scatter(area_1stage.flatten()[gain_1stage_mask], gbw_final)
plt.yscale("log")
plt.xscale("log")
```
